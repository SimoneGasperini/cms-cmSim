```python
import pandas as pd

dtier = 'AOD'
df = pd.read_parquet(f'./../data/{dtier.lower()}_data_history.parquet')
df.head(None)
```

```python
import pylab as plt
from cmSim.container import DataContainer

container = DataContainer.from_dataframe(df)

fig, ax = plt.subplots(figsize=(16, 12))
container.plot_storage_history_by_datatier(
    ax, datatiers=[dtier, dtier+'SIM'], tier='T1', norm=False)
```

```python
fig, ax = plt.subplots(figsize=(16, 12))
container.plot_storage_history_by_datatier(
    ax, datatiers=[dtier, dtier+'SIM'], tier='T2', norm=False)
```

```python
from cmSim import utils

sim = df[df['tier'] == dtier+'SIM']
container = DataContainer.from_dataframe(df=sim)

pags = utils.get_pags()

fig, ax = plt.subplots(figsize=(16, 12))
container.plot_storage_history_by_pag(ax, pags=pags, tier='T1', norm=True)
```
