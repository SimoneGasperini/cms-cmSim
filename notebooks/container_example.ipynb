```python
import pandas as pd

df = pd.read_parquet('./../data/aod_data_history.parquet')
df.head(None)
```

```python
import pylab as plt

fig, ax = plt.subplots(figsize=(16, 12))
container.plot_storage_history_in_sites(ax, tier='T2', norm=True)
```
