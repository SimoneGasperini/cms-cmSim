```python
import pandas as pd
from collections import Counter

filepath = './../data/dataset_site_info.parquet'
df = pd.read_parquet(filepath)
```

```python
def get_process(name):
    string = name.replace('/', ' ').replace('_', ' ')
    process = string.split()[0]
    return process
```

```python
df['process'] = df['dataset_name'].apply(get_process)
```

```python
df_aod = df[df['tier'] == 'AOD']

print('number of distinct AOD datasets =', df_aod['dataset_name'].nunique())
print('number of distinct AOD physics processes =', df_aod['process'].nunique())

df_aod.head()
```

```python
df_aodsim = df[df['tier'] == 'AODSIM']

print('number of distinct AODSIM datasets =', df_aodsim['dataset_name'].nunique())
print('number of distinct AODSIM physics processes =', df_aodsim['process'].nunique())

df_aodsim.head()
```

```python
process_count_aod = Counter(df_aod.drop_duplicates(subset=['dataset_name'])['process'])

process_count_aod.most_common(20)
```

```python
process_count_aodsim = Counter(df_aodsim.drop_duplicates(subset=['dataset_name'])['process'])

process_count_aodsim.most_common(20)
```
